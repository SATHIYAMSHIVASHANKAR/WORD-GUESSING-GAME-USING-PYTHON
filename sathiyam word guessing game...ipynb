```python
import random  # library that we use in order to choose on random words from a list of words

name = input("HEY Wellcome !!!! What is your name? ") # Here the variable collects the name of the user
 
print("Lets have fun :-), Good Luck ! ", name)
 
words = ['chocolate', 'baskin robbins', 'icecream', 'ibaco',
         'codeing', 'kfc', 'subway', 'dominos',
         'macdonals', 'water', 'pizzahut', 'sathiya']
 
word = random.choice(words) # A rondom word is choosed form the list
 
print("~~~~~~ Guess the characters ~~~~~~")
 
guesses = ''
 
turns = 12 # number of guessing turns
 

while turns > 0:
     
    failed = 0  # counts the number of times a user fails
     
    for char in word: # calculates the guesses made by user
         
        if char in guesses: # compares the character in guesses with the randomly selected word.
            print(char)
             
        else:
            print("_")
             
            failed += 1 # for every failure 1 incremented in failure is done.       
 
    if failed == 0:

        print("hurry !!!!! You Win :-)") # user will win the game if failure is 0 and 'You Win' is displayed
        
        print("The random word is : ", word)
        break
     
    guess = input("OOPS !! guess a character:") # if worng input is eentered, it askes for another guess.
     
    guesses += guess # every input character will be stored in guesses
     
    if guess not in word: # This conditional looks for the worng guessed charector from the user.
         
        turns -= 1
         
        print("OOPS !! Wrong guess") # if the character doesn’t match the word then “Wrong” will be given as output
         
        # this will print the number of
        # turns left for the user
        print("You have with", + turns, 'more guesses')
         
         
        if turns == 0:
            print("That Was a Good try :-) better luck next time !! You Loose :-(")
```
